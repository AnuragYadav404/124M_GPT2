```python
with open('./dataset/input.txt', 'r', encoding='utf-8') as f:
    text = f.read()

# print(text)
```

```python
# next we need to create a tokenizer
# the tokenizer will create a vocabulary for the input data
# we will be using the tiktoken tokenizer, which is a byte pair encoding (BPE) tokenizer
import tiktoken
# here we are using a pre-trained tokenizer that is compatible with the GPT-2 model
# benefit of using a pre-trained tokenizer is that it will have a larger vocabulary and will be able to encode more words
# we can also train our own tokenizer, as that better suits our specific dataset
# a bpe tokenizer will break down words into subwords, which allows it to handle out-of-vocabulary words better
# the way a bpe 
# how is a simple tokenizer trained?
# the most simple form of tokenizer is a character-level tokenizer, which simply splits the text into individual characters
# the then trained tokenizer can then be used to encode the text into a sequence of integers, which can then be fed into the model
# so when we use a pre-trained tokenizer, we are using a tokenizer that has already been trained on a large corpus of text, and has a fixed vocabulary
# the tiktoken library provides a simple interface for encoding and decoding text using the pre-trained tokenizer
tokenizer = tiktoken.get_encoding('gpt2')
# we can then use the tokenizer to encode the text into a sequence of integers
encoded_text = tokenizer.encode(text)
print(encoded_text[:100])  # print the first 100 tokens
```

```python
# so now we have the classic question of tensors:
# the data, should be living as on a tensor on GPU
# we can use the PyTorch library to create a tensor from the encoded text
import torch
# we can create a tensor from the encoded text
# we need to specify the data type of the tensor, which is usually torch.long for integer
encoded_text_tensor = torch.tensor(encoded_text, dtype=torch.long)
print(encoded_text_tensor[:100])  # print the first 100 tokens as a tensor
# but we also need to move the tensor to the GPU, so that it can be used for training the model
# but initially let's keep the data on the CPU, and we can move it to the GPU later when we are ready to train the model
# what benefit does tensor provide over a simple list of integers?
# tensors are optimized for performance and can be easily moved to the GPU for faster computation, lists of integers are not optimized for performance and cannot be easily moved to the GPU
   
```

```python
block_size = 8
batch_size = 4
def get_batch(split=None):
    # lets first build the most simple one:
    # here we get only a single xb, and yb set
    # we first need to know the length of our context: block_size # defined as 8
    # we need a starting index 
        # conditions for starting index: random
        # sample space: 
        # n is data length, say we n = 100
        # start idx, can be 0 ... 100-block_size-1, -1 is because xy, is idx+1
    # why do we need a split?
    # split so that we take data based on training data or validation data

    start_idx = torch.randint(len(data)-block_size-1, (batch_size,))
    print(start_idx)
    x = torch.stack([data[idx:idx+block_size] for idx in start_idx])
    y = torch.stack([data[idx+1:idx+1+block_size] for idx in start_idx])
    return x,y

get_batch()
```
